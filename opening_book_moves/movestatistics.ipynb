```python
import json
import matplotlib.pyplot as plt
from collections import Counter, OrderedDict

with open('move9.json', 'r') as file:
    data = json.load(file)

move_counts = sorted( Counter(data).items(), key=lambda x: x[1], reverse=True)
moves = [move for move, freq in move_counts] 
frequencies = [freq for move, freq in move_counts]
opening_book_list = []
for move in OrderedDict((move, freq) for move, freq in zip(moves[:40], frequencies[:40])):
    opening_book_list.append(move)
print(opening_book_list)
move1 = moves[0:20]
move2 = moves[20:50]
frequencies1 = frequencies[0:20]
frequencies2 = frequencies[20:50]

plt.figure(figsize=(12, 6))
plt.bar(move1, frequencies1, color='skyblue')
plt.xlabel('Moves')
plt.ylabel('Frequency')
plt.title('Frequency of Each Move')
plt.xticks(rotation=90) 
plt.grid(axis='y')
plt.tight_layout()
plt.show()
plt.figure(figsize=(12, 6))
plt.bar(move2, frequencies2, color='skyblue')
plt.xlabel('Moves')
plt.ylabel('Frequency')
plt.title('Frequency of Each Move')
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()


```
