```python
import json
import matplotlib.pyplot as plt
from collections import Counter

with open('move9.json', 'r') as file:
    data = json.load(file)


# Counting the frequency of each move (key)
move_counts = Counter(data)
# Sorting the move counts by frequency in descending order
move_counts = sorted(move_counts.items(), key=lambda x: x[1], reverse=True)

# Extracting the moves and their frequencies
moves = [move for move, freq in move_counts]  # Get the moves
frequencies = [freq for move, freq in move_counts]  # Get the frequencies


# split moves and frequencies into four lists
move1 = moves[0:20]
move2 = moves[20:50]
move3 = moves[50:100]
move4 = moves[100:]
frequencies1 = frequencies[0:20]
frequencies2 = frequencies[20:50]
frequencies3 = frequencies[50:100]
frequencies4 = frequencies[100:]





# Plotting the bar chart
plt.figure(figsize=(12, 6))
plt.bar(move1, frequencies1, color='skyblue')
plt.xlabel('Moves')
plt.ylabel('Frequency')
plt.title('Frequency of Each Move')
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()

# Plotting the bar chart
plt.figure(figsize=(12, 6))
plt.bar(move2, frequencies2, color='skyblue')
plt.xlabel('Moves')
plt.ylabel('Frequency')
plt.title('Frequency of Each Move')
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()

# Plotting the bar chart
plt.figure(figsize=(12, 6))
plt.bar(move3, frequencies3, color='skyblue')
plt.xlabel('Moves')
plt.ylabel('Frequency')
plt.title('Frequency of Each Move')
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()

# Plotting the bar chart
plt.figure(figsize=(12, 6))
plt.bar(move4, frequencies4, color='skyblue')
plt.xlabel('Moves')
plt.ylabel('Frequency')
plt.title('Frequency of Each Move')
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```
